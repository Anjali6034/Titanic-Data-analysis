```python
#Loading and Understanding the Data
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df=sb.load_dataset('titanic') #load the dataset from seaborn
```

```python
#Display first 5 rows
df.head() #Shows first 5 rows to get an overview of data
```

```python
# Check data types and missing values
df.info() # shows the data type of the columns and non-null counts.
```

```python
df.describe() #summary of the data.
# give count, mean, std, min, max, etc.
```

```python
df.isnull().sum() #show the total missing values in every column
```

```python
# Print rows where 'deck' is missing
missing_age_rows = df[df['deck'].isnull()]
display(missing_age_rows)
```

# handling the missing values


```python
 #fill missing age value with median
df['age'].fillna(df['age'].median())
```

```python
df.info()
```

```python
#fill the missing value in embarked with mode.
df['embarked'].fillna(df['embarked'].mode()[0])
```

```python
df.info() 
```

```python
missing_deck_ratio = df['deck'].isnull().sum() / len(df)
print(missing_deck_ratio)
if missing_deck_ratio > 0.5:
    df.drop(columns=['deck'], inplace=True)
     # Drop 'deck' if more than 50% missing
    print("'deck' column dropped due to excessive missing values.")
else:
    df['deck'].fillna('Unknown', inplace=True)
     # Otherwise, fill missing values with 'Unknown'
```

```python
df.info()
```

## halndling the categorical variables


```python
# One-hot encode Embarked column
df = pd.get_dummies(df, columns=['embarked'], drop_first=True)
```

```python
# Convert Pclass to categorical
df['pclass'] = df['pclass'].astype('category')
```

```python
df.info()
```

# univariate analysis

```python
#histogram for age
sb.histplot(df['age'],bins=30,kde=True)
plt.title('age distribution')
plt.show()
```

```python
#bar plot for survival count.
sb.countplot(x='survived',data=df, color='green')
plt.title('survival count')
plt.show()
```

```python
#countplot for sex
data=df.copy()
sb.countplot(x='sex',data=data, color='magenta')
plt.title('gender distribution')
plt.xticks([0,1],['male','female'])
plt.show()
```

# bivariate analysis

```python
#survival rate by sex
sb.barplot(x='sex',y='survived',data=df, color='orange')
plt.title('survival rate by gender')
plt.xticks([0,1],['male','female'])
plt.show()
```

```python
#boxplot oof age by survival
sb.boxplot(x='survived',y='age',data=df, color='brown')
plt.title('age distribution by survival')
plt.show() 
```

```python
#select only numeric oclumns for correlation matrix
numeric_df=df.select_dtypes(include=['number'])
print(numeric_df)
#compute correlation matrix
x=numeric_df.corr()
#heatmap of correlation
plt.figure(figsize=(10,6))
sb.heatmap(x,annot=True,cmap='coolwarm')
plt.title('feature correlation heatmap')
plt.show()
```

```python
# Outlier Detection
# Boxplot for Fare
sb.boxplot(x=df['fare'])
plt.title("Fare Outliers")
plt.show()
```

```python
# Pie Chart for Survival
survived_counts = df['survived'].value_counts()
plt.pie(survived_counts, labels=['Not Survived', 'Survived'], autopct='%1.1f%%', startangle=90)
plt.title("Survival Rate")
plt.show()
```

```python
# Pairplot
sb.pairplot(df[['age', 'fare', 'survived']], hue='survived')
plt.show()
```
